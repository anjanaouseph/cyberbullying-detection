Creating BoW

```python
import nltk
nltk.download('punkt')
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
def tokenize(text):
  tokens = nltk.word_tokenize(text)
  return tokens

# Replace 'your_file.csv' with the actual file path
file_path = 'https://raw.githubusercontent.com/vaamps/cyberbullying-detection/main/datasets/output_sentiment.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)
# Replace 'your_text_column' with the actual text column name
text_column = df['Text']
# vocab = pd.read_csv('filtered_vocab.txt', header=None)[0].tolist()

import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')

# Specify the path to your vocab file
import requests

# URL of the vocab file
vocab_file_path = 'https://raw.githubusercontent.com/vaamps/cyberbullying-detection/main/datasets/negative_words.txt'

# Read the vocab file and create a set of non-stopwords
response = requests.get(vocab_file_path)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    vocab_words = {word.strip() for word in response.text.split() if word.strip()}
    print(vocab_words)
else:
    print(f"Failed to fetch the vocabulary file. Status code: {response.status_code}")

# Remove stopwords and create a new vocabulary
filtered_vocabulary = set(word.lower() for word in vocab_words if word.lower() not in stopwords.words('english'))

# Create a TfidfVectorizer
vectorizer = TfidfVectorizer(tokenizer=tokenize, vocabulary=filtered_vocabulary)

# Fit and transform the text column to get the BoW features
bow_features = vectorizer.fit_transform(text_column)

# Convert the BoW features to a DataFrame
#bow_df = pd.DataFrame(bow_features.toarray(), columns=vectorizer.get_feature_names_out())

# Get the feature names
feature_names = vectorizer.get_feature_names_out()


# Create a DataFrame using the CSR matrix
bow_df = pd.DataFrame.sparse.from_spmatrix(bow_features, columns=feature_names)


# Concatenate the original DataFrame with the new BoW DataFrame
df = pd.concat([df, bow_df], axis=1)
df
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.svm import SVC
from sklearn.feature_selection import SelectKBest, mutual_info_classif
from sklearn.metrics import accuracy_score  # Import accuracy_score
import numpy as np

# Assuming X (features) and y (target labels) are already defined

# Create an SVM classifier
svm_classifier = SVC()

# Number of features to select
k_best = 100

# Define the number of folds for cross-validation
k_folds = 3

# Create a Stratified K-Fold cross-validator
stratified_kfold = StratifiedKFold(n_splits=k_folds, shuffle=True, random_state=42)

# Perform K-Fold Cross-Validation for SVM
training_scores = []
validation_scores = []

for train_index, test_index in stratified_kfold.split(X, y):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    # Feature Selection
    feature_selector = SelectKBest(mutual_info_classif, k=k_best)
    X_train_selected = feature_selector.fit_transform(X_train, y_train)
    X_test_selected = feature_selector.transform(X_test)

    # Fit the SVM classifier
    svm_classifier.fit(X_train_selected, y_train)

    # Predictions and scoring on the training set
    y_train_pred = svm_classifier.predict(X_train_selected)
    training_score = accuracy_score(y_train, y_train_pred)
    training_scores.append(training_score)

    # Predictions and scoring on the validation set
    y_val_pred = svm_classifier.predict(X_test_selected)
    validation_score = accuracy_score(y_test, y_val_pred)
    validation_scores.append(validation_score)

# Calculate average training and validation scores
avg_training_score = np.mean(training_scores)
avg_validation_score = np.mean(validation_scores)

print(f"Average Training Accuracy: {avg_training_score}")
print(f"Average Validation Accuracy: {avg_validation_score}")
```

```python
from sklearn.model_selection import StratifiedKFold, RandomizedSearchCV
from sklearn.svm import SVC
from sklearn.feature_selection import SelectKBest, mutual_info_classif
from sklearn.metrics import accuracy_score
import numpy as np

# Assuming X (features) and y (target labels) are already defined

# Define the parameter grid for SVM
param_grid = {
    'C': np.logspace(-4, 4, 20),  # Regularization parameter
    'gamma': np.logspace(-4, 4, 20),  # Kernel coefficient
    'kernel': ['rbf', 'linear', 'poly', 'sigmoid']  # Kernel type
}

# Create an SVM classifier instance
svm_classifier = SVC()

# Perform randomized search with cross-validation
random_search = RandomizedSearchCV(svm_classifier, param_grid, n_iter=10, cv=5, random_state=42)
random_search.fit(X, y)

# Print the best hyperparameters
print("Best Hyperparameters:", random_search.best_params_)

# Use the best hyperparameters for k-fold cross-validation
best_params = random_search.best_params_
k_folds = 3
stratified_kfold = StratifiedKFold(n_splits=k_folds, shuffle=True, random_state=42)
training_scores = []
validation_scores = []

for train_index, test_index in stratified_kfold.split(X, y):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

    # Feature Selection
    feature_selector = SelectKBest(mutual_info_classif, k=100)
    X_train_selected = feature_selector.fit_transform(X_train, y_train)
    X_test_selected = feature_selector.transform(X_test)

    # Fit the SVM classifier with best hyperparameters
    svm_classifier = SVC(C=best_params['C'], gamma=best_params['gamma'], kernel=best_params['kernel'])
    svm_classifier.fit(X_train_selected, y_train)

    # Calculate accuracy
    y_train_pred = svm_classifier.predict(X_train_selected)
    y_val_pred = svm_classifier.predict(X_test_selected)
    training_scores.append(accuracy_score(y_train, y_train_pred))
    validation_scores.append(accuracy_score(y_test, y_val_pred))

# Calculate average training and validation scores
avg_training_score = np.mean(training_scores)
avg_validation_score = np.mean(validation_scores)

print(f"Average Training Accuracy: {avg_training_score}")
print(f"Average Validation Accuracy: {avg_validation_score}")
```
