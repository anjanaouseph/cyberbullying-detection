Creating BoW

```python
import nltk
nltk.download('punkt')
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
def tokenize(text):
  tokens = nltk.word_tokenize(text)
  return tokens

# Replace 'your_file.csv' with the actual file path
file_path = 'https://raw.githubusercontent.com/vaamps/cyberbullying-detection/main/datasets/output_sentiment.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)
# Replace 'your_text_column' with the actual text column name
text_column = df['Text']
# vocab = pd.read_csv('filtered_vocab.txt', header=None)[0].tolist()

import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')

# Specify the path to your vocab file
import requests

# URL of the vocab file
vocab_file_path = 'https://raw.githubusercontent.com/vaamps/cyberbullying-detection/main/datasets/filtered_vocab.txt'

# Read the vocab file and create a set of non-stopwords
response = requests.get(vocab_file_path)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    vocab_words = {word.strip() for word in response.text.split() if word.strip()}
    print(vocab_words)
else:
    print(f"Failed to fetch the vocabulary file. Status code: {response.status_code}")

# Remove stopwords and create a new vocabulary
filtered_vocabulary = set(word.lower() for word in vocab_words if word.lower() not in stopwords.words('english'))

# Create a TfidfVectorizer object
vectorizer = TfidfVectorizer(max_features=1000, min_df=5, max_df=0.85, ngram_range=(1, 2))

# Fit and transform the text column to get the BoW features
bow_features = vectorizer.fit_transform(text_column)

# Convert the BoW features to a DataFrame
#bow_df = pd.DataFrame(bow_features.toarray(), columns=vectorizer.get_feature_names_out())

# Get the feature names
feature_names = vectorizer.get_feature_names_out()


# Create a DataFrame using the CSR matrix
bow_df = pd.DataFrame.sparse.from_spmatrix(bow_features, columns=feature_names)


# Concatenate the original DataFrame with the new BoW DataFrame
df = pd.concat([df, bow_df], axis=1)
df
```

Setting X and y

```python
# df.to_csv('output_bow.csv')
new_df = df.iloc[:, 5:]
new_df.head()
```

```python
# All imports for task 2
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score, KFold
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
X = new_df
y = df['oh_label']
# X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2)
print(X.shape[0])
print(y.shape[0])
```

Running Gradient Boost ML model (with k-cross)

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import RandomizedSearchCV
import numpy as np

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

# Define the Gradient Boosting Classifier
gb_clf = GradientBoostingClassifier()

# Specify the parameter grid
param_grid = {
    'n_estimators': np.arange(50, 201, 50),  # Adjust the range as needed
    'learning_rate': [0.01, 0.05, 0.1, 0.2, 0.3],
    'max_depth': [3, 4, 5, 6],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Create RandomizedSearchCV
random_search = RandomizedSearchCV(gb_clf, param_distributions=param_grid, n_iter=5, cv=3, random_state=42)

# Fit the model
random_search.fit(X_train, y_train)

# Print the best parameters
print("Best Parameters:", random_search.best_params_)

# Evaluate on the test set
accuracy = random_search.score(X_test, y_test)
print("Test Accuracy:", accuracy)
```
